## Removing Duplicated data

```python
df=df.drop_duplicates(subset=["Recipe Name", "Ingredients" ,"Link"])
df
```

## Handling Missing Values

```python
df = df.dropna(subset=["Recipe Name", "Ingredients" ,"Link"])
df
```

```python
df.isnull().sum()
```

## Verifying links 

```python
df = df[df['Link'].str.startswith('http',na=False)].reset_index(drop=True)
```

```python
df
```

## Converting Recipe name and ingredients to Lower Case

```python
df['Recipe Name'] = df['Recipe Name'].str.lower().str.strip()
df['Ingredients'] = df['Ingredients'].str.lower().str.strip()
```

```python
df
```

## Removing non english words

```python
import re
```

```python
df['Recipe Name'] = df['Recipe Name'].apply(lambda x:' '.join(re.findall(r'[a-zA-Z\s]+',x)))
```

```python
df['Ingredients'] = df['Ingredients'].apply(lambda x:' '.join(re.findall(r'[a-zA-Z,\s]+',x)))
```

```python
df = df[df['Ingredients'].str.strip().apply(lambda x: re.sub(r',', '', x).strip() != '')]

df = df.reset_index(drop=True)
df = df.drop(index=87)
df.iloc[87]
```

```python
df.to_csv('cleaned_recipes.csv', index=False)
```
